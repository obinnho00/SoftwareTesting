```python
import unittest
```

```python
class Triangle:
    
    def classify_triangle(self,a, b, c):
        if a + b <= c or a + c <= b or b + c <= a:
            return "Not a triangle"
        elif a == b == c:
            return "Equilateral"
        elif a == b or b == c or a == c:
            return "Isosceles"
        else:
            return "Scalene"   
            
class TestTriangleClassification(unittest.TestCase):

    # Test cases for Equilateral Triangle
    def test_equilateral_triangle(self):
        m = Triangle()
        self.assertEqual(m.classify_triangle(5, 5, 5), "Equilateral")  # Valid Equilateral
        self.assertEqual(m.classify_triangle(5, 5, 6), "Isosceles")     # Boundary: Near Equilateral, but Isosceles

    # Test cases for Isosceles Triangle
    def test_isosceles_triangle(self):
        m = Triangle()
        self.assertEqual(m.classify_triangle(5, 5, 4), "Isosceles")    # Valid Isosceles
        self.assertEqual(m.classify_triangle(5, 5, 6), "Isosceles")    # Valid Isosceles
        self.assertEqual(m.classify_triangle(5, 6, 6), "Isosceles")    # Valid Isosceles
        self.assertEqual(m.classify_triangle(5, 5, 10), "Not a triangle")  # Boundary: Invalid, a + b = c

    # Test cases for Scalene Triangle
    def test_scalene_triangle(self):
        m = Triangle()
        self.assertEqual(m.classify_triangle(5, 6, 7), "Scalene")      # Valid Scalene
        self.assertEqual(m.classify_triangle(5, 6, 12), "Not a triangle")  # Boundary: Invalid, a + b < c

    # Test cases for Invalid Triangles
    def test_invalid_triangle(self):
        m = Triangle()
        self.assertEqual(m.classify_triangle(1, 2, 3), "Not a triangle")    # Boundary: a + b = c
        self.assertEqual(m.classify_triangle(1, 10, 12), "Not a triangle")  # Boundary: a + b < c
        self.assertEqual(m.classify_triangle(5, 5, 10), "Not a triangle")   # Boundary: a + b = c



if __name__ == "__main__":
    unittest.main(argv=[''], verbosity=2, exit=False)
```

```python
class Grade:

    def get_grade(self, score):
        if score < 0 or score > 100:
            return "Invalid score"
        elif score >= 90:
            return "A"
        elif score >= 80:
            return "B"
        elif score >= 70:
            return "C"
        elif score >= 60:
            return "D"
        else:
            return "F"
            
class Test_Grade(unittest.TestCase):

    # Test case for A grade
    def test_Agrade(self):
        g = Grade()
        self.assertEqual(g.get_grade(90), "A")
        self.assertEqual(g.get_grade(99), "A")
        self.assertEqual(g.get_grade(100), "A")
        self.assertEqual(g.get_grade(101), "Invalid score")

    # Test case for B grade
    def test_Bgrade(self):
        g = Grade()
        self.assertEqual(g.get_grade(80), "B")  # min boundary for B
        self.assertEqual(g.get_grade(89), "B")  # max boundary for B

    # Test case for C grade
    def test_Cgrade(self):
        g = Grade()
        self.assertEqual(g.get_grade(70), "C")  # min boundary for C
        self.assertEqual(g.get_grade(79), "C")  # max boundary for C

    # Test case for D grade
    def test_Dgrade(self):
        g = Grade()
        self.assertEqual(g.get_grade(60), "D")  # min boundary for D
        self.assertEqual(g.get_grade(69), "D")  # max boundary for D

    # Test case for F grade
    def test_Fgrade(self):
        g = Grade()
        self.assertEqual(g.get_grade(0), "F")
        self.assertEqual(g.get_grade(59), "F")  # max boundary for F

    # Test cases for invalid scores
    def test_invalid(self):
        g = Grade()
        self.assertEqual(g.get_grade(-1), "Invalid score")  # less than 0
        self.assertEqual(g.get_grade(101), "Invalid score")  # greater than 100


if __name__ == "__main__":
    unittest.main(argv=[''], verbosity=2, exit=False)
```

```python
class Eligibility:

    def check_eligibility(self, age, income, credit_score, loan_amount):
        if age < 18 or income < 25000 or credit_score < 600 or credit_score > 850 or loan_amount > (income * 0.5):
            return "Denied"
        return "Approved"

class TestForEligibility(unittest.TestCase):

    def test_valid_cases(self):
        e = Eligibility()
      
        self.assertEqual(e.check_eligibility(18, 25000, 600, 12500), "Approved")  # Valid boundary values
        self.assertEqual(e.check_eligibility(30, 50000, 700, 25000), "Approved")  # Typical valid case

    def test_age_boundary(self):
        e = Eligibility()
        # Age less than 18 should be denied
        self.assertEqual(e.check_eligibility(17, 30000, 700, 15000), "Denied")  # Invalid age
        # Age exactly 18 should be approved
        self.assertEqual(e.check_eligibility(18, 30000, 700, 15000), "Approved")  # Valid age boundary

    def test_income_boundary(self):
        e = Eligibility()
        # Income less than 25000 should be denied
        self.assertEqual(e.check_eligibility(25, 24999, 700, 10000), "Denied")  # Invalid income boundary
        # Income exactly 25000 should be approved
        self.assertEqual(e.check_eligibility(25, 25000, 700, 12000), "Approved")  # Valid income boundary

    def test_credit_score_boundary(self):
        e = Eligibility()
        # Credit score less than 600 should be denied
        self.assertEqual(e.check_eligibility(30, 30000, 599, 15000), "Denied")  # Invalid credit score (low)
        # Credit score exactly 600 should be approved
        self.assertEqual(e.check_eligibility(30, 30000, 600, 15000), "Approved")  # Valid credit score (lower boundary)
        # Credit score more than 850 should be denied
        self.assertEqual(e.check_eligibility(30, 30000, 851, 15000), "Denied")  # Invalid credit score (high)

    def test_loan_amount_boundary(self):
        e = Eligibility()
        # Loan amount greater than 50% of income should be denied
        self.assertEqual(e.check_eligibility(25, 30000, 700, 16000), "Denied")  # Loan exceeds 50% of income
        # Loan amount exactly 50% of income should be approved
        self.assertEqual(e.check_eligibility(25, 30000, 700, 15000), "Approved")  # Valid loan boundary

    def test_invalid_cases(self):
        e = Eligibility()
        # Combination of invalid values
        self.assertEqual(e.check_eligibility(17, 24000, 500, 20000), "Denied")  # Invalid age, income, credit score
        self.assertEqual(e.check_eligibility(19, 30000, 601, 20000), "Denied")  # Loan amount exceeds limit

if __name__ == "__main__":
    unittest.main(argv=[''], verbosity=2, exit=False)
```
